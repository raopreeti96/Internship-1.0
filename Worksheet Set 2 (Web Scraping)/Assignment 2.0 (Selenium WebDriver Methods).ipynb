# Xpath Syntax


Xpath knows as the xml path is a language that helps to query the xml documents . In consist of expression for a path along with certain conditions to locate a particular element .


The below format of xpath is mentioned below.









# Types of XPaths


- These are two types of XPaths

   - Absolute XPath
   
   - Relative XPath
   

# Absolute XPath - :

Begins with the single forword slasg (/), means we select the element from the root node and go all the way down to the element neended.


/html/body/div[2]/div[1]/div/h4[1]/b/html[1]/body[1]/div[2]/h4[1]/b[1]




# Relative XPath -:

Starts from the middle of the HTML Structure. Start with double forword slash (//) and can search elements anywhere on the webpage without the long absolute xpath.

//dib[@class='featured-box columnsize1']//h4[1]//b[1]


# XPath Function

1 - Basic XPath : XPath expression select nodes or list of nodes on the basis of attributes like ID , Name , Classname etc. From the xml document like shows below.


XPath = //input[@name='uid]


xPath = //a[@href = "http://google.com/']



# Contains () - : 

Used when the value of any attribute changes dynamically , example , login information . In can find element with partial text.


For finding 'Submit' button where type ='Submit'


XPath =//*[contains(@type,'sub)]

XPatch = //img[contains(@src,'content')]




# Using OR and AND ->

Here , two conditions are used , wheather 1st condition or 2nd condition should be true . Means any one condition should be true to find the element .


XPath =//*[@type='Submit' or @ name = 'btnReset']


In AND expressionn , two conditions are used , both conditions should be true to find the element . it fails to find element if anyone condition is false .


XPatch =//input[@type='submit'and@name ='btnlogin']



# Start-With () -: 


Used to find a web element whose value of an attribute changes on the refresh or on any other  dynamic operation on the web page we match the starting text of the attribute to locate an element whose attribute has changed dynamically.



XPath = //imh[starts-with (@sec ,/'https']



# text() :-

Used with the text function to locate an element with exact text.


Here it go anywhere inside the document , irrespective of the tag , but , it must contain a text whose value is search google or type a URL . the asterisk (*) implies any tag with some value .


XPath =//*[text()='search Google or type a URL']

# Q9: Write a python program to scrape data for Top 1000 Quotes of All Time.

2. Click on Top Quotes


3. Than scrap 

   
   a) Quote

   c) Type Of Quote

```python
url="https://www.azquotes.com/" # Here i am assigning the link in url namea variable 
driver.get(url) #here using the get method to access the link
```

```python
#Clicking using absolute xpath function
top_quotes=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
top_quotes.click()
```

```python
#Lets find 1000 Quotes
quote=[]

start=0
end=10

                         
for page in range(start,end):
    Quote=driver.find_elements(By.XPATH,"//a[@class='title']")
    
    for i in Quote:
        quote.append(i.text)
    
    
print(len(quote))


```

```python
#Lets find author name
author=[]

start=0
end=10

                         
for page in range(start,end):
    Author=driver.find_elements(By.XPATH,"//div[@class='wrap-block']")
    
    for i in Author:
        author.append(i.text.split('\n')[1])
    
    
    
print(len(author))

```

```python
#Lets find Time of Quote
top=[]

start=0
end=10

                         
for page in range(start,end):
    TOP=driver.find_elements(By.XPATH,"//div[@class='tags']")
    
    for i in TOP:
        top.append(i.text)
    
    
print(len(top))

```

```python
#Lets MakeDataFrame
import pandas as pd
thousand_quote=pd.DataFrame({"Best Quote":quote,"Best Author":author,"Topic of Quote":top})
thousand_quote
```
